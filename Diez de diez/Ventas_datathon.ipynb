```python
import pandas as pd
import os
import matplotlib
import numpy as np
```

```python
ventas_csv= pd.read_csv("sample_data/Venta.csv")
meta_csv= pd.read_csv("sample_data/Meta_venta.csv")
train_csv= pd.read_csv("sample_data/DIM_TIENDA.csv")
test_csv= pd.read_csv("sample_data/DIM_TIENDA_TEST.csv")
train= pd.DataFrame(train_csv)
test= pd.DataFrame(test_csv)
ventas= pd.DataFrame(ventas_csv)
meta= pd.DataFrame(meta_csv)
```

```python
Complete_set= pd.concat([train, test])
Complete_set.head(20)
```

```python
columnas_to_ventas= pd.DataFrame(Complete_set, columns=["TIENDA_ID", "ENTORNO_DES"])
columnas_to_ventas.head()
```

```python
ventas= pd.merge(ventas, columnas_to_ventas, on="TIENDA_ID", how='inner')
ventas.head()
```

```python
ventas = pd.merge(ventas, meta, on='ENTORNO_DES', how='left')
ventas.head()
```

```python
ventas.to_csv("ventas_ent.csv", index=False)
```
