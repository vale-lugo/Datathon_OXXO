```python
# Definir el rango de metros cuadrados (ajusta según tus datos si conoces el rango real)
mts_min = df_clean['MTS2VENTAS_NUM'].min()
mts_max = df_clean['MTS2VENTAS_NUM'].max()

# Función para estimar puertas por tamaño
def estimar_puertas(row):
    if row['PUERTASREFRIG_NUM'] == 0:
        # Escalar proporcionalmente entre 6 y 18 puertas
        proporcion = (row['MTS2VENTAS_NUM'] - mts_min) / (mts_max - mts_min)
        estimado = 6 + proporcion * (18 - 6)
        return round(estimado)
    else:
        return row['PUERTASREFRIG_NUM']

# Aplicar
df_clean['PUERTASREFRIG_NUM'] = df_clean.apply(estimar_puertas, axis=1)
```

# Competencias

```python
import pandas as pd

# Cargar los archivos CSV en DataFrames
# Specify the encoding as 'latin-1' to handle characters not compatible with utf-8
df1 = pd.read_csv('denue_inegi_46111_.csv', encoding='latin-1')
df2 = pd.read_csv('denue_inegi_46112-46311_.csv', encoding='latin-1')
df3 = pd.read_csv('denue_inegi_46321-46531_.csv', encoding='latin-1')
df4 = pd.read_csv('denue_inegi_46591-46911_.csv', encoding='latin-1')


# Definir los rangos de latitud y longitud
lat_min, lat_max = 24.84057, 26.40643
lon_min, lon_max = -100.44299, -97.94741

# Función para filtrar un DataFrame según coordenadas
def filtrar_por_coordenadas(df, lat_col='latitud', lon_col='longitud'):
    return df[
        (df[lat_col] >= lat_min) & (df[lat_col] <= lat_max) &
        (df[lon_col] >= lon_min) & (df[lon_col] <= lon_max)
    ]

# Aplicar el filtrado a cada DataFrame
df1_filtrado = filtrar_por_coordenadas(df1)
df2_filtrado = filtrar_por_coordenadas(df2)
df3_filtrado = filtrar_por_coordenadas(df3)
df4_filtrado = filtrar_por_coordenadas(df4)

# Unir los 4 DataFrames filtrados
df_total = pd.concat([df1_filtrado, df2_filtrado, df3_filtrado, df4_filtrado], ignore_index=True)

# Guardar resultado a CSV si lo deseas
#df_total.to_csv('tiendas_nl_tam.csv', index=False)

# Mostrar número de tiendas encontradas
print(f'Total de tiendas en el rango especificado: {len(df_total)}')
```

```python
# Filtrar filas que NO contienen "oxxo" (sin importar mayúsculas/minúsculas)
df_filtrado_sin_oxxo = df_total[~df_total['nom_estab'].str.contains('oxxo', case=False, na=False)]

# Guardar resultado a CSV si lo deseas
#df_filtrado_sin_oxxo.to_csv('tiendas_nl_tam_sin_oxxo.csv', index=False)

# Mostrar el número de tiendas restantes
print(f'Tiendas sin "oxxo": {len(df_filtrado_sin_oxxo)}')
```

```python
palabras_clave_competencia = [
    "abarrotes", "tienda", "super", "mini", "minisuper",
    "six", "extra", "express", "kiosko", "circulo k",
    "7 eleven", "7-eleven", "seven", "deposito",
    "conveniencia", "gasolinera", "oxxo"
]
```

```python
import re

# Creamos una expresión regular que combina todas las palabras clave
regex = '|'.join(palabras_clave_competencia)

# Filtramos solo las tiendas que contienen alguna palabra clave en el nombre
df_competencia_relevante = df_filtrado_sin_oxxo[
    df_filtrado_sin_oxxo['nom_estab'].str.contains(regex, case=False, na=False)
]
```

```python
print("Tiendas competencia posibles:", len(df_competencia_relevante))
```

```python
import pandas as pd

# Crear una lista para guardar el conteo de competencia
competencia_500 = []

# Rango de búsqueda en grados (aproximadamente 500 m)
delta = 0.0045

# Iterar sobre cada tienda OXXO
for idx, tienda in df_clean.iterrows():
    lat = tienda['LATITUD_NUM']
    lon = tienda['LONGITUD_NUM']

    # Filtrar las tiendas de competencia dentro del rango
    tiendas_cercanas = df_competencia_relevante[
        (df_filtrado_sin_oxxo['latitud'] >= lat - delta) &
        (df_filtrado_sin_oxxo['latitud'] <= lat + delta) &
        (df_filtrado_sin_oxxo['longitud'] >= lon - delta) &
        (df_filtrado_sin_oxxo['longitud'] <= lon + delta)
    ]

    # Guardar el número de tiendas cercanas
    competencia_500.append(len(tiendas_cercanas))

# Añadir la nueva columna al DataFrame
df_clean['COMPETENCIA_500'] = competencia_500

# Guardar resultado si lo deseas
#df_clean.to_csv('oxxos_con_competencia.csv', index=False)

# Mostrar una muestra
print(df_clean[['LATITUD_NUM', 'LONGITUD_NUM', 'COMPETENCIA_500']].head())
```
